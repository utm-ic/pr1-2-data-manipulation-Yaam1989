## Paso 3: Limpieza de Datos

Se revisaron los datos en busca de valores nulos o duplicados. Se decidió qué hacer con ellos, ya sea eliminarlos, reemplazarlos o aplicar otra estrategia. También se verificó que los tipos de datos en cada columna fueran correctos y se convirtieron aquellos que no estaban en el formato adecuado.

```python
# Calcula el total de valores faltantes
total_missing = df.isnull().sum().sum()

# Calcula el total de valores en el dataset
total_values = df.size

# Calcula el porcentaje
missing_percentage = (total_missing / total_values) * 100

print(f"Porcentaje total de valores faltantes: {missing_percentage:.2f}%")
```

```python
# Muestra nombres de las columnas con valores faltantes y su cantidad
missing_values_by_column = df.isnull().sum()
print(missing_values_by_column[missing_values_by_column > 0])
```

```python
#Se creó una copia del dataset para realizar las modificaciones correspondientes
df_copy = df.copy()
categorical_cols = ['Medication_Use','Substance_Use']  # Lista de columnas categóricas
df_copy[categorical_cols] = df_copy[categorical_cols].fillna('Never')

```

```python
# Esa linea comprueba que ya no existan valores nulos
df_copy.isnull().sum()
```

## Paso 5: Filtrado de Datos

Se identificaron subconjuntos específicos del dataset. Se filtraron las personas con un puntaje alto en ansiedad (mayor a 80) y se clasificaron según las horas de sueño reportadas (menos de 5 horas, entre 5 y 7 horas, y más de 7 horas). Posteriormente, se analizó la variación del nivel de estrés en cada grupo.



```python
# Define los rangos para la clasificación
bins_anxiety = [0, 6, 15, 20]
labels_anxiety = ['Low', 'Normal', 'High']

df_copy['Anxiety_Score_Level'] = pd.cut(df_copy['Anxiety_Score'], bins=bins_anxiety, labels=labels_anxiety)

# Cuenta usuarios en cada nivel y calcula porcentajes
nivel_counts = df_copy['Anxiety_Score_Level'].value_counts().sort_index()
total = nivel_counts.sum()
nivel_percentages = (nivel_counts / total) * 100

# Grafica
plt.figure(figsize=(6,4))
ax = sns.barplot(x=nivel_percentages.index, y=nivel_percentages.values, palette="viridis")

# Añade etiquetas con el porcentaje exacto
for i, pct in enumerate(nivel_percentages.values):
    ax.text(i, pct + 0.5, f"{pct:.1f}%", ha='center', fontsize=12, fontweight='bold')

# Visualización
plt.xlabel("Nivel", fontsize=12)
plt.ylabel("Porcentaje de usuarios", fontsize=12)
plt.title("Usuarios por nivel de ansiedad", fontsize=14, fontweight='bold')
plt.ylim(0, 100)  # Ajustar el eje Y para que no se corte

plt.show()
```

```python

# Define los rangos para la clasificación

bins_sleep = [0, 5, 8, 13]  # Ejemplo de rangos para Sleep_Hours
labels_sleep = ['Low', 'Normal', 'High']

df_copy['Sleep_Hours_Level'] = pd.cut(df_copy['Sleep_Hours'], bins=bins_sleep, labels=labels_sleep)

# Cuenta usuarios en cada nivel y calcula porcentajes
nivel_counts = df_copy['Sleep_Hours_Level'].value_counts().sort_index()
total = nivel_counts.sum()
nivel_percentages = (nivel_counts / total) * 100

# Grafica
plt.figure(figsize=(6,4))
ax = sns.barplot(x=nivel_percentages.index, y=nivel_percentages.values, palette="viridis")

# Añade etiquetas con el porcentaje exacto
for i, pct in enumerate(nivel_percentages.values):
    ax.text(i, pct + 0.5, f"{pct:.1f}%", ha='center', fontsize=12, fontweight='bold')

# Visualización
plt.xlabel("Nivel", fontsize=12)
plt.ylabel("Porcentaje de usuarios", fontsize=12)
plt.title("Usuarios por Horas de sueño", fontsize=14, fontweight='bold')
plt.ylim(0, 100)  # Ajustar el eje Y para que no se corte

plt.show()
```

## Paso 7: Guardado de Resultados

Se filtró el dataset para encontrar a las personas que recibían terapia y que además tenían un nivel bajo de estrés (por debajo de la mediana). Una vez obtenido este subconjunto, se guardó la información en un nuevo archivo CSV para futuros análisis.

```python
import pandas as pd

# Calcular la mediana del nivel de estrés
stress_median = df_copy["Stress_Level"].median()

# Filtrar personas que reciben terapia (1) y tienen un nivel de estrés bajo
filtered_df = df_copy[(df_copy["Therapy"] == 1) & (df_copy["Stress_Level"] < stress_median)]

# Guardar el subconjunto en un nuevo archivo CSV
filtered_df.to_csv("baja_terapia_estres.csv", index=False)

# Mostrar cuántas personas cumplen con ambas condiciones
print("Número de personas que reciben terapia y tienen estrés bajo:", len(filtered_df))
```
